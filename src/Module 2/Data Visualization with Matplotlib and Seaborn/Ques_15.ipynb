```python
# Seaborn Plot - Heatmap
# Objective: Create a heatmap to visualize a correlation matrix with Seaborn, using
# advanced styling.
```

```python
# Step 1: Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Step 2: Load dataset (using seaborn's built-in dataset for this example)
# For your case, you can replace this with your own dataset
df = sns.load_dataset('iris')

# Step 3: Compute the correlation matrix
correlation_matrix = df.corr()

# Step 4: Create the heatmap with advanced styling
plt.figure(figsize=(10, 8))  # Increase the figure size for better readability
sns.set(style='white')  # White background style
heatmap = sns.heatmap(correlation_matrix, 
                    annot=True,  # Add annotations with correlation values
                    cmap='coolwarm',  # Color palette for the heatmap
                    fmt='.2f',  # Format for displaying the numbers
                    linewidths=0.5,  # Line width for separating cells
                    linecolor='gray',  # Color of the grid lines
                    cbar_kws={'shrink': 0.8},  # Shrink the color bar for better visualization
                    vmin=-1, vmax=1)  # Define the range of color scale (-1 to 1 for correlation)

# Step 5: Improve axis labels and add titles
plt.title('Correlation Matrix Heatmap', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=12)  # Rotate x-axis labels for better readability
plt.yticks(rotation=0, fontsize=12)  # Keep y-axis labels horizontal

# Show the heatmap
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Step 1: Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Step 2: Load dataset (using seaborn's built-in dataset for this example)
df = sns.load_dataset('iris')

# Step 3: Exclude non-numeric columns and compute the correlation matrix
numeric_df = df.select_dtypes(include=[np.number])  # Select only numeric columns
correlation_matrix = numeric_df.corr()  # Compute the correlation matrix

# Step 4: Create the heatmap with advanced styling
plt.figure(figsize=(10, 8))  # Increase the figure size for better readability
sns.set(style='white')  # White background style
heatmap = sns.heatmap(correlation_matrix, 
                    annot=True,  # Add annotations with correlation values
                    cmap='coolwarm',  # Color palette for the heatmap
                    fmt='.2f',  # Format for displaying the numbers
                    linewidths=0.5,  # Line width for separating cells
                    linecolor='gray',  # Color of the grid lines
                    cbar_kws={'shrink': 0.8},  # Shrink the color bar for better visualization
                    vmin=-1, vmax=1)  # Define the range of color scale (-1 to 1 for correlation)

# Step 5: Improve axis labels and add titles
plt.title('Correlation Matrix Heatmap', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=12)  # Rotate x-axis labels for better readability
plt.yticks(rotation=0, fontsize=12)  # Keep y-axis labels horizontal

# Show the heatmap
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
