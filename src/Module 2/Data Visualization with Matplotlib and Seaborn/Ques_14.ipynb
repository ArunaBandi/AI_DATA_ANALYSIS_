```python
# Styling with Matplotlib - Line Plot
# Objective: Create a line plot with customized styles using Matplotlib.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a line plot with customized styles
plt.plot(x, y, 
        color='b',           # Blue line
        linestyle='--',       # Dashed line
        marker='o',          # Circle markers at data points
        markersize=6,        # Size of markers
        markerfacecolor='r', # Red colored markers
        markeredgewidth=2,   # Border width of the markers
        linewidth=2)         # Width of the line

# Adding labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Styled Line Plot')

# Display grid
plt.grid(True)

# Show plot
plt.show()
```
