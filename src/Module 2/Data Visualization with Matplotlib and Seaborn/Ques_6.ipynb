```python
# Customizing Matplotlib Plot Styles
# Description: Customize the Matplotlib bar plot from Question 1 with a grid, different color, and custom fonts.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]

# Set custom font (ensure it's installed on your system)
plt.rcParams['font.family'] = 'serif'  # You can change to 'Comic Sans MS', 'Arial', etc.

# Create the plot
fig, ax = plt.subplots()

# Bar plot with custom color
bars = ax.bar(categories, values, color='skyblue', edgecolor='black')

# Add grid
ax.grid(True, linestyle='--', linewidth=0.5, axis='y', alpha=0.7)

# Add title and labels with custom font sizes
ax.set_title('Custom Bar Plot', fontsize=16, fontweight='bold')
ax.set_xlabel('Category', fontsize=12)
ax.set_ylabel('Values', fontsize=12)

# Add value labels above bars
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2.0, yval + 1, int(yval), ha='center', va='bottom', fontsize=10)

# Show plot
plt.tight_layout()
plt.show()
```
