```python
# Advanced Scatter Plot with Seaborn
# Description: Using Seaborn, enhance the scatter plot for height vs. weight by adding size for points based
# on data: Size [10, 15, 13, 18, 12].
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'Height': [150, 160, 170, 180, 190],
    'Weight': [50, 60, 70, 80, 90],
    'Size': [10, 15, 13, 18, 12]  # This will control point size
}

# Create DataFrame
df = pd.DataFrame(data)

# Create scatter plot
plt.figure(figsize=(8, 6))
scatter = sns.scatterplot(data=df, x='Height', y='Weight', size='Size', sizes=(50, 500), legend=False)

# Optional enhancements
plt.title('Advanced Scatter Plot: Height vs Weight with Variable Size')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)

plt.show()
```
