```python
# Combining Plots with Matplotlib
# Objective: Combine multiple plots into a single figure using Matplotlib's subplots.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Create a figure with 2 rows and 2 columns of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot data on the first subplot (top-left)
axs[0, 0].plot(x, y1, label='sin(x)')
axs[0, 0].set_title('Sine Function')
axs[0, 0].legend()

# Plot data on the second subplot (top-right)
axs[0, 1].plot(x, y2, label='cos(x)', color='orange')
axs[0, 1].set_title('Cosine Function')
axs[0, 1].legend()

# Plot data on the third subplot (bottom-left)
axs[1, 0].plot(x, y3, label='tan(x)', color='green')
axs[1, 0].set_title('Tangent Function')
axs[1, 0].legend()
axs[1, 0].set_ylim(-10, 10)  # Limit the y-axis to avoid extreme values of tan(x)

# Remove the fourth subplot (bottom-right)
fig.delaxes(axs[1, 1])

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
