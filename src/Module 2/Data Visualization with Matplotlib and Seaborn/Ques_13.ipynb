```python
# Customizing Plots with Seaborn - Boxplot
# Objective: Customize a boxplot with Seaborn to show the distribution of petal lengths by
# species.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset('iris')

# Create the boxplot
plt.figure(figsize=(8, 6))  # Adjust figure size for better visibility
sns.boxplot(x='species', y='petal_length', data=iris, palette='Set2')

# Customize the plot
plt.title('Distribution of Petal Lengths by Species', fontsize=14)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Petal Length (cm)', fontsize=12)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset('iris')

# Create the boxplot with hue for species
plt.figure(figsize=(8, 6))  # Adjust figure size for better visibility
sns.boxplot(x='species', y='petal_length', data=iris, hue='species', palette='Set2', dodge=False)

# Customize the plot
plt.title('Distribution of Petal Lengths by Species', fontsize=14)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Petal Length (cm)', fontsize=12)

# Show the plot
plt.show()
```
