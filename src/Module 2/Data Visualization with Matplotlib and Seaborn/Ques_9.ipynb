```python
# Matplotlib Subplots for Multiple Data Sets
# Description: Create subplots using Matplotlib to display the bar plot for product sales and the scatter plot for
# height vs. weight side by side.
```

```python
import matplotlib.pyplot as plt

# Sample data for product sales
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales = [120, 150, 90, 200, 130]

# Sample data for height vs weight
heights = [150, 160, 170, 180, 190, 200]
weights = [50, 60, 70, 80, 90, 100]

# Create a figure and axis for subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Bar plot for product sales (first subplot)
axs[0].bar(products, sales, color='skyblue')
axs[0].set_title('Product Sales')
axs[0].set_xlabel('Products')
axs[0].set_ylabel('Sales')

# Scatter plot for height vs weight (second subplot)
axs[1].scatter(heights, weights, color='green')
axs[1].set_title('Height vs Weight')
axs[1].set_xlabel('Height (cm)')
axs[1].set_ylabel('Weight (kg)')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
```
