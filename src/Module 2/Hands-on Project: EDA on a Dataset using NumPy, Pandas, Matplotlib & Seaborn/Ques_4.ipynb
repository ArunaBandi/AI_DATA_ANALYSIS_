```python
# Title: Basic Data Visualization
# Description: Plot the distribution of one feature using Seaborn.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data: Let's say we have a dataset with a 'feature' column
import numpy as np
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)  # 1000 random numbers from a normal distribution

# Plot the distribution
sns.displot(data, kde=True)  # kde=True adds a kernel density estimate curve to the plot

# Set the title and labels
plt.title('Distribution of the Feature')
plt.xlabel('Feature Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
